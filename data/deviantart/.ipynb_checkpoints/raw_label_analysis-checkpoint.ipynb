```python
import glob
import os
import string

import itertools
import enchant
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import wordninja

from chars import CharDict
from words import WordDict

pd.set_option('display.max_rows', 4000)
%load_ext autoreload
%autoreload 2
```

```python
with open("metadata.txt", 'r') as f:
    lines = f.readlines()
image_files = glob.glob("images/*")
print("# of labels:", len(lines))
print("# of image files:", len(image_files))
```

```python
def lookup_dictionary(s, dictionary=WordDict):
    ## 1 : In dictionary, include it. Do not split
    ## 0 : In dictionary but exclude it.
    ## -1 : Not in dictionary.
    
    if s == '':
        return 0
    
    is_english = WordDict.enchantment.check(s)
    
    if s in dictionary.INCLUDING_WORDS or is_english and s not in WordDict.EXCLUDING_WORDS:
        return 1
    elif s in dictionary.EXCLUDING_WORDS:
        return 0
    else:
        return -1

def split_scriptiocontinua(s, dictionary=WordDict):
    lookup_result = lookup_dictionary(s, dictionary=dictionary)
    if lookup_result == 1:
        return s
    elif lookup_result == 0:
        return ''
    else:
        words = wordninja.split(s)
        if all(map(lambda s: lookup_dictionary(s, dictionary=dictionary) == 1, words)):
            return ' '.join(words)
        else:
            return ''
    
def handle_numerals(s, dictionary=WordDict):
    if s in dictionary.INCLUDING_WORDS:
        return s
    else:
        for num in range(10):
            s = s.replace(str(num), '')
        return s

def is_fully_printable(s):
    return all([i in string.printable for i in s])

def replace_chars(s, dictionary=CharDict):
    ## Replace punctuation
    s = "".join(list(map(lambda x: dictionary.PUNCTUATION_REPLACEMENT_DICT.get(x, x), s)))
    
    ## Replace letters
    s = "".join(list(map(lambda x: dictionary.LETTER_REPLACEMENTS_DICT.get(x, x), s)))
    return s
```

```python
l = ['21', '90s', 'germy', 'giant', 'la', 'legendary', 'pink', 'pokémon', 'pokemon', 
     'regis', 'registeel', 'semaine', 'steel', 'week', 'legendaire', 'pokegraph', 
     'germy21', 'pinkgermy', 'de', '']
print(l)
l = list(map(replace_chars, l))
l = list(map(handle_numerals, l))
l = list(map(split_scriptiocontinua, l))
l = list(filter(lambda s: s != '', l))
l = list(set(l))
print(l)
```

```python
tag_collection = []
tag_collection_org = []
for i, line in enumerate(lines):
    print(i)
    row = line.strip().split(',')
    image_file = row[0]
    tags_org = row[1:].copy()
    tag_collection_org.extend(tags_org)
    
    ## Replace bad chars
    tags = list(map(replace_chars, tags_org))
    
    ## Lowercase
    tags = list(map(lambda s: s.lower(), tags))
    
    ## Remove words with non-printable chars
    tags = list(filter(is_fully_printable, tags))
    
    ## Remove unwanted words
    tags = list(filter(lambda s: s not in WordDict.EXCLUDING_WORDS, tags))
    
    ## Handle numerals
    tags = list(map(handle_numerals, tags))
    
    ## Wordninja
    tags = list(map(split_scriptiocontinua, tags))
    
    ## Remove unwanted words again
    tags = list(filter(lambda s: s not in WordDict.EXCLUDING_WORDS, tags))

    ## Remove empty strings
    tags = list(filter(lambda s: s != '', tags))
    
    ## Make set
    tags = list(set(tags))
    
    ## Add tags to collection
    tag_collection.extend(tags)
    
#     print(image_file)
#     print('\t', sorted(tags_org))
#     print()
#     print('\t', sorted(tags))
#     print('-------------------------------------------------------------------')
    
# tag_collection = list(set(tag_collection))
```

```python
counts = pd.value_counts(tag_collection_org)
for tag_org in sorted(list(set(map(lambda s: s.strip(), tag_collection_org)))):

    tag_org_org = tag_org

    ## Replace bad chars
    tag_org = replace_chars(tag_org)

    ## Lowercase
    tag_org = tag_org.lower()
    
    ## Remove words with non-printable chars
    if not is_fully_printable(tag_org):
        tag = '<REMOVED>'
    else:
        tag = tag_org
        ## Remove unwanted words
        if tag not in WordDict.EXCLUDING_WORDS:
            
            if tag in WordDict.MAPPING_WORDS:
                tag = WordDict.MAPPING_WORDS[tag]
            else:
                ## Handle numerals
                tag = handle_numerals(tag)
                
                if tag in WordDict.MAPPING_WORDS:
                    tag = WordDict.MAPPING_WORDS[tag]
                else:
                    ## Wordninja
                    tag = split_scriptiocontinua(tag)
        else:
            tag = '<EXCLUDED>'
            
    ## Remove unwanted words again
    if tag in WordDict.EXCLUDING_WORDS:
        tag = '<EXCLUDED>'

    print("'{}' --------> '{}' ({})".format(tag_org, tag, counts.loc[tag_org_org]))
```
