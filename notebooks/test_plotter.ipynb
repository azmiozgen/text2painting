```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
TRAIN_CSV_FILE = "../models/united_af25796d7170/united_train_log.csv"
VAL_CSV_FILE = "../models/united_af25796d7170/united_val_log.csv"
```

```python
df_train = pd.read_csv(TRAIN_CSV_FILE, delimiter=',')
df_val = pd.read_csv(VAL_CSV_FILE, delimiter=',')
```

```python
df_train
```

```python
df_val
```

```python
df = df_train
l = len(df.index)
xticks = 10
plt.figure(figsize=(15, 8))
plt.suptitle("Training losses")

plt.subplot(1, 2, 1)
plt.title("Generator loss")
plt.xlabel("Iterations")
plt.ylabel("Loss")

plt.grid()
plt.yticks(np.arange(min(df['G_loss']), max(df['G_loss']), 2.0))

plt.plot(df['Iteration'], df['G_loss'])

plt.subplot(1, 2, 2)
plt.title("Discriminator loss")
plt.xlabel("Iterations")
plt.ylabel("Loss")

plt.grid()
plt.yticks(np.arange(min(df['D_loss']), max(df['D_loss']), 0.1))

plt.plot(df['Iteration'], df['D_loss'])
```
